### New PATHS TO FMRI FILES

```python
import os
from glob import glob
import pickle

# Get list of all directories inside the target folder
directories = glob("/scratch/leuven/369/vsc36935/NEO_demo/*/", recursive=True)

postpath = "rfMRI_REST_hp0_clean.nii.gz"

# Create correct full paths to the desired files
for i in range(len(directories)):
    subject_id = os.path.basename(os.path.normpath(directories[i]))  # e.g., HCA8983208_V1_MR
    directories[i] = os.path.join(directories[i], postpath)

# Save the list of paths to a pickle file
with open("paths_to_nifti.pkl", "wb") as fp:
    pickle.dump(directories, fp)

print(f"Saved {len(directories)} paths to paths_to_nifti.pkl")
```

```python
# Replace with your actual .pkl file path
file_path = 'paths_to_nifti.pkl'

# Open and load the pickle file
with open(file_path, 'rb') as f:
    data = pickle.load(f)

# Display the contents
print("Loaded data from pickle file:")
print(data)
```
